```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats, special
from scipy.optimize import minimize_scalar
from mpl_toolkits.mplot3d import Axes3D
import warnings
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
# 设置中文字体支持
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置支持中文的字体，优先使用SimHei
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
sns.set_theme(style="white", font_scale=1.2)

df = pd.read_excel('filtered-ref-converted.xlsx')
df_low = df[df['孕妇BMI'] < 27].copy()
df_mid = df[(df['孕妇BMI'] >= 27) & (df['孕妇BMI'] <= 38)].copy()
df_high = df[df['孕妇BMI'] > 37.75].copy()
```

```python
def prepare_data(df, cols):
    """
    输入：df，cols（三元组列名）
    输出：cleaned numpy array (n x d) 和原始 df_clean
    """
    df_clean = df[list(cols)].copy()
    df_clean = df_clean.dropna().reset_index(drop=True)
    if df_clean.shape[0] < 30:
        print("警告：样本量较小（{}），t-copula 参数估计可能不稳定。".format(df_clean.shape[0]))
    return df_clean.values, df_clean

# 2. 边缘分布：ECDF（经验分布）与逆经验分位函数
# ---------------------------
def ecdf_u(arr):
    """
    返回 arr 对应的经验分布转换到 (0,1) 的 u 值（使用 rank/(n+1)）
    """
    n = len(arr)
    ranks = stats.rankdata(arr, method='average')  # 1..n
    u = ranks / (n + 1.0)
    return u

def empirical_inverse_quantile(u, arr_sorted):
    """
    根据排序好的样本 arr_sorted，把 u (0,1) 映射回原始尺度（线性插值）
    u: array
    arr_sorted: sorted 1d array of original data
    """
    n = len(arr_sorted)
    # sample quantile positions
    pos = np.linspace(1/(n+1), n/(n+1), n)
    return np.interp(u, pos, arr_sorted)

# 3. t-Copula 的对数似然（给定 nu，通过样本 x = t^{-1}(u;nu) 来估计 R）
# ---------------------------
def t_copula_loglik(nu, U):
    """
    nu: degrees of freedom (float)
    U: n x d matrix of uniforms (0,1)
    返回：-loglik（因为后续用 minimize）
    算法思想：
      - 对每个维度，把 u -> x = t_ppf(u; nu)
      - 用 x 的样本相关矩阵作为 R 的估计
      - 计算 t-copula 密度 c(u) = f_d(x; R, nu) / prod_j f_1(x_j; nu)
      - sum log c(u)
    """
    n, d = U.shape
    # 边界数值处理
    eps = 1e-10
    U = np.clip(U, eps, 1-eps)
    # 量化到 t 分位
    X = stats.t.ppf(U, df=nu)  # n x d
    # 样本协方差 -> 相关矩阵
    # 先中心化
    Xc = X - X.mean(axis=0)
    # 获得协方差矩阵（无偏）
    S = np.cov(Xc, rowvar=False, bias=False)
    # 计算相关矩阵 R from S
    D = np.sqrt(np.diag(S))
    outerD = np.outer(D, D)
    # 防止数值问题
    with np.errstate(divide='ignore', invalid='ignore'):
        R = S / outerD
    # 修正对角
    R[np.isnan(R)] = 0.0
    np.fill_diagonal(R, 1.0)
    # 确保正定：若不是，则进行最近似正定矩阵修正（简单版：小量加对角）
    try:
        # try cholesky to test PD
        np.linalg.cholesky(R)
    except np.linalg.LinAlgError:
        # add tiny jitter
        jitter = 1e-6
        R = R + np.eye(d) * jitter
        # if still fails, increase jitter
        k = 0
        while True:
            try:
                np.linalg.cholesky(R)
                break
            except np.linalg.LinAlgError:
                jitter *= 10
                R = R + np.eye(d) * jitter
                k += 1
                if k > 10:
                    raise RuntimeError("无法使相关矩阵正定")
    # 计算常数项
    # multivariate t pdf constant
    log_c_terms = []
    # precompute inverse and logdet
    R_inv = np.linalg.inv(R)
    sign, logdet = np.linalg.slogdet(R)
    if sign <= 0:
        # fallback
        logdet = np.log(np.abs(np.linalg.det(R)) + 1e-12)

    # Precompute gamma/log constants
    lgamma_half_nu_d = special.gammaln((nu + d)/2.0)
    lgamma_half_nu = special.gammaln(nu/2.0)
    const_multivar = lgamma_half_nu_d - lgamma_half_nu - 0.5* (d * np.log(nu * np.pi) + logdet)
    const_univar = special.gammaln((nu+1)/2.0) - special.gammaln(nu/2.0) - 0.5*np.log(nu*np.pi)

    # For each sample compute log density ratio
    quad = np.einsum('ij,jk,ik->i', X, R_inv, X)  # x^T R^{-1} x for each row
    # multivar log pdf
    log_f_d = const_multivar - ((nu + d)/2.0) * np.log(1.0 + quad/nu)
    # univariate log pdfs sum
    log_f_1_sum = np.sum(const_univar - ((nu+1)/2.0) * np.log(1.0 + (X**2)/nu), axis=1)
    # copula log density
    log_c = log_f_d - log_f_1_sum
    total_loglik = np.sum(log_c)
    # return negative for minimization
    return -total_loglik, R, X

def negloglik_for_nu(nu, U):
    val, _, _ = t_copula_loglik(nu, U)
    return val


def fit_t_copula(U, nu_bounds=(2.01, 200.0)):
    """
    输入：U (n x d)
    输出：dict {nu, R, loglik, X (t-quantiles)}
    """
    # 优化 nu（单变量实数优化），使用 minimize_scalar
    res = minimize_scalar(lambda v: negloglik_for_nu(v, U), bounds=nu_bounds, method='bounded', options={'xatol':1e-3})
    nu_hat = res.x
    negloglik_val, R_hat, X_hat = t_copula_loglik(nu_hat, U)
    loglik = -negloglik_val
    return {'nu': float(nu_hat), 'R': R_hat, 'loglik': float(loglik), 'X': X_hat, 'success': res.success}


# 5. 从拟合的 t-copula 生成样本（仿真）
# ---------------------------
def simulate_from_t_copula(n_samples, nu, R):
    """
    生成 n_samples 个 t-copula 的 U(0,1) 样本
    步骤：
      - draw z ~ MVN(0, R)
      - draw chi2 ~ Chi2(nu)
      - x = z * sqrt(nu / chi2)
      - u = t.cdf(x; nu)
    """
    d = R.shape[0]
    # Cholesky
    L = np.linalg.cholesky(R)
    Z = np.random.normal(size=(n_samples, d))
    # correlate
    Z_corr = Z @ L.T
    chi2 = np.random.chisquare(nu, size=n_samples)
    X = Z_corr * np.sqrt(nu / chi2)[:, None]
    U = stats.t.cdf(X, df=nu)
    return U, X
```

```python
def run_t_copula_pipeline(df_high,
                          cols = ("孕周","孕妇BMI","Y染色体浓度"),
                          do_plot=True,
                          sim_n=1000):
    # prepare
    data_array, df_clean = prepare_data(df_high, cols)
    n, d = data_array.shape
    print("样本数：{}, 维度：{}".format(n, d))

    # PIT：经验分布转化为 U
    U = np.zeros_like(data_array, dtype=float)
    for j in range(d):
        U[:, j] = ecdf_u(data_array[:, j])

    # fit t-copula
    print("开始拟合 t-copula（优化自由度 nu）...")
    fitres = fit_t_copula(U)
    print("拟合完成： nu = {:.3f}, loglik = {:.3f}, success={}".format(fitres['nu'], fitres['loglik'], fitres['success']))
    print("相关矩阵 R_hat =\n", np.round(fitres['R'], 4))

    # 模拟并映射回原始尺度
    U_sim, X_sim = simulate_from_t_copula(sim_n, fitres['nu'], fitres['R'])
    # 将 U_sim 逆映射回原始尺度（用经验逆分位）
    arr_sorted_list = [np.sort(data_array[:, j]) for j in range(d)]
    sim_orig = np.zeros_like(U_sim)
    for j in range(d):
        sim_orig[:, j] = empirical_inverse_quantile(U_sim[:, j], arr_sorted_list[j])

    # plot 原始 vs 模拟（3D scatter）
    if do_plot:
        fig = plt.figure(figsize=(12,5))
        ax1 = fig.add_subplot(1,2,1, projection='3d')
        ax1.scatter(data_array[:,0], data_array[:,1], data_array[:,2], s=10)
        ax1.set_title("Original Samples ({} points)".format(n))
        ax1.set_xlabel("Gestational Age (weeks)")
        ax1.set_ylabel("Maternal BMI")
        ax1.set_zlabel("Y-chromosome conc.")
    
        ax2 = fig.add_subplot(1,2,2, projection='3d')
        ax2.scatter(sim_orig[:,0], sim_orig[:,1], sim_orig[:,2], s=6, alpha=0.6)
        ax2.set_title("t-copula Simulation ({} points)".format(sim_n))
        ax2.set_xlabel("Gestational Age (weeks)")
        ax2.set_ylabel("Maternal BMI")
        ax2.set_zlabel("Y-chromosome conc.")
    
        plt.tight_layout()
        plt.show()

    return {'fit': fitres, 'simulated_original_scale': sim_orig, 'U': U, 'data_array': data_array}
```

```python
import numpy as np
import pandas as pd
from tqdm import trange

def bootstrap_t_copula(df_high, cols=("孕周","孕妇BMI","Y染色体浓度"), B=500):
    data_array, df_clean = prepare_data(df_high, cols)
    n, d = data_array.shape
    R_boot = np.zeros((B, d, d))
    nu_list = []

    for b in trange(B):
        # 重抽样
        idx = np.random.choice(n, n, replace=True)
        sample = data_array[idx, :]

        # 转换成 U
        U = np.zeros_like(sample, dtype=float)
        for j in range(d):
            U[:, j] = ecdf_u(sample[:, j])

        # 拟合 t-copula
        try:
            fitres = fit_t_copula(U)
            R_boot[b] = fitres['R']
            nu_list.append(fitres['nu'])
        except Exception:
            R_boot[b] = np.nan
    
    return np.array(R_boot), np.array(nu_list)


# 运行 Bootstrap
R_boot, nu_list = bootstrap_t_copula(df_high, B=500)

# 计算每个相关系数的统计量
mean_R = np.nanmean(R_boot, axis=0)
std_R = np.nanstd(R_boot, axis=0)
low_R = np.nanpercentile(R_boot, 2.5, axis=0)
high_R = np.nanpercentile(R_boot, 97.5, axis=0)

print("Bootstrap 相关矩阵均值：\n", np.round(mean_R, 3))
print("Bootstrap 相关矩阵标准差：\n", np.round(std_R, 3))
print("Bootstrap 95% CI 下界：\n", np.round(low_R, 3))
print("Bootstrap 95% CI 上界：\n", np.round(high_R, 3))

print("自由度 nu 的均值和95%CI: {:.2f} [{:.2f}, {:.2f}]".format(
    np.nanmean(nu_list), 
    np.nanpercentile(nu_list, 2.5), 
    np.nanpercentile(nu_list, 97.5)
))
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline

# ------------- helper: empirical inverse quantile (already in your pipeline) -------------
def empirical_inverse_quantile(u, arr_sorted):
    n = len(arr_sorted)
    pos = np.linspace(1/(n+1), n/(n+1), n)
    return np.interp(u, pos, arr_sorted)

# ------------- conditional sampling from fitted t-copula -------------
def conditional_sample_y_given_t_b(u1, u2, fitres, n_draw=2000):
    """
    在给定 U1=u1 (孕周 的边缘cdf), U2=u2 (BMI 的边缘cdf) 下，
    从拟合好的 t-copula (fitres['nu'], fitres['R']) 条件抽取 U3 的样本。
    返回 U3_samples (n_draw,)
    方法：利用多元 t 的条件分布性质（近似用蒙特卡洛 rejection-free 方法）：
      - 生成大量 (U1,U2,U3) 的联合样本（来自 t-copula），筛选最接近给定 (u1,u2) 的子样本，
        并取对应的 U3 作为条件样本（实用且直观，样本越多越精确）。
    说明：这是 Monte Carlo 条件模拟的简易实现（不显式求解析条件分布），
          在维度很低时效果良好。
    """
    nu = fitres['nu']
    R = fitres['R']
    d = R.shape[0]
    assert d == 3, "本实现假定三维 copula：顺序应为 (孕周, BMI, Y浓度)"

    # 先生成一大批联合 U 样本
    n_pool = max(20000, n_draw*50)  # 池大小，可调整
    # 生成 t-copula 样本 (U)
    # 生成 correlated t-samples via MVN + chi2
    L = np.linalg.cholesky(R)
    Z = np.random.normal(size=(n_pool, d))
    Z_corr = Z @ L.T
    chi2 = np.random.chisquare(nu, size=n_pool)
    X = Z_corr * np.sqrt(nu / chi2)[:, None]
    U_pool = stats.t.cdf(X, df=nu)  # n_pool x d

    # 选择在 (u1,u2) 附近的点：按欧氏距离在 U-space 中选择最近的 n_draw*factor 个，然后按权重（kernel）重采样
    target = np.array([u1, u2])
    pool12 = U_pool[:, :2]
    dists = np.linalg.norm(pool12 - target[None, :], axis=1)
    # 选 topK
    K = int(n_draw * 6)  # 取的池大小（6 倍冗余），可调整
    K = min(max(K, n_draw), n_pool)
    idx_sorted = np.argsort(dists)[:K]
    selected = U_pool[idx_sorted]  # K x 3
    # 用距离做权重（kernel），距离越小权重越大
    eps = 1e-9
    weights = np.exp(- (dists[idx_sorted]**2) / (2*(0.02**2) + eps))  # bandwidth 可调整（0.02）
    weights = np.maximum(weights, 1e-12)
    weights = weights / weights.sum()
    # 按权重重采样得到 n_draw 个 U3
    choose_idx = np.random.choice(np.arange(K), size=n_draw, replace=True, p=weights)
    U3_samples = selected[choose_idx, 2]
    return U3_samples

# ------------- compute prob(t | bmi) via conditional sim and empirical marginals -------------
def prob_y_ge_threshold_given_t_b(t_val, b_val, fitres, data_array, arr_sorted_list, y_thresh,
                                  n_cond_draw=2000):
    """
    输入:
      - t_val: 孕周值（实际数值）
      - b_val: BMI 值（实际数值）
      - fitres: t-copula 拟合结果字典，包含 'nu' 和 'R'
      - data_array: 原始 n x 3 数据，列顺序：(孕周, BMI, Yconc)
      - arr_sorted_list: 每列排序后的样本列表（用于经验逆分位）
      - y_thresh: 阈值（原尺度）
    返回: prob = P(Y >= y_thresh | t=t_val, BMI=b_val)
    """
    # 1) map t_val, b_val to their empirical CDF (u1, u2) using original sample
    #    用 rank/(n+1) 的插值
    n = data_array.shape[0]
    # empirical cdf via interpolation
    def empirical_cdf(val, arr_sorted):
        pos = np.linspace(1/(n+1), n/(n+1), n)
        return np.interp(val, arr_sorted, pos, left=pos[0], right=pos[-1])
    u1 = empirical_cdf(t_val, arr_sorted_list[0])
    u2 = empirical_cdf(b_val, arr_sorted_list[1])
    # 防边界
    u1 = np.clip(u1, 1e-6, 1-1e-6)
    u2 = np.clip(u2, 1e-6, 1-1e-6)

    # 2) conditional sample U3 | U1=u1, U2=u2
    U3_samps = conditional_sample_y_given_t_b(u1, u2, fitres, n_draw=n_cond_draw)

    # 3) map U3 -> Y samples via empirical inverse quantile on Y 的排序样本
    y_sim = empirical_inverse_quantile(U3_samps, arr_sorted_list[2])

    # 4) compute prob
    prob = np.mean(y_sim >= y_thresh)
    return prob, y_sim

# ------------- cost function C(t) as you provided -------------
def cost_C(t):
    if t <= 12:
        C = 0.1
    elif 12 < t <= 27:
        C = (0.1 / np.exp(12)) * np.exp(t-12) + 0.1
    else:
        C = 1000.0
    return C

# ------------- grid search optimizer for best t -------------
def find_optimal_t(fitres, data_array, arr_sorted_list,
                   y_thresh, lam=1.0,
                   bmi_fixed=None, bmi_grid=None,
                   t_grid=np.linspace(10,27,71),   # 默认 10~27 周，步长 ~0.25 周
                   n_cond_draw=2000,
                   maximize=True):
    """
    - 如果提供 bmi_fixed，则在该 BMI 下求最佳 t。
    - 如果提供 bmi_grid（可为 array 或 None），则对 bmi_grid 的每个 BMI 求最佳 t 并返回。
    - 如果两者都不提供，默认对 data_array 中 BMI 的中位数进行优化。
    返回 dict 包含 grid 结果和推荐值。
    """
    if (bmi_fixed is None) and (bmi_grid is None):
        bmi_fixed = np.median(data_array[:,1])

    results = []
    for t in t_grid:
        if bmi_grid is None:
            b_vals = [bmi_fixed]
        else:
            b_vals = list(bmi_grid)

        # 我们支持对多个 BMI 取平均效用（期望）
        U_vals = []
        prob_vals = []
        for b in b_vals:
            prob, _ = prob_y_ge_threshold_given_t_b(t, b, fitres, data_array, arr_sorted_list, y_thresh,
                                                   n_cond_draw=n_cond_draw)
            C = cost_C(t)
            U = prob - lam * C
            U_vals.append(U)
            prob_vals.append(prob)
        # 如果是多 BMI（bmi_grid），以平均 prob/U 作为整体效用
        mean_U = np.mean(U_vals)
        mean_prob = np.mean(prob_vals)
        results.append({'t':t, 'mean_U':mean_U, 'mean_prob':mean_prob})

    # 选择最优
    arr = results
    if maximize:
        best = max(arr, key=lambda x: x['mean_U'])
    else:
        best = min(arr, key=lambda x: x['mean_U'])
    return {'grid_results':results, 'best':best}

# ---------------- example usage ----------------
# 假设你已经有：
#   fitres (包含 'nu' 和 'R'), data_array (n x 3), arr_sorted_list
# 你只需设置 y_thresh, lam, 和你关心的 BMI（或 bmi_grid）
#
# 例子（可直接运行）：

res = run_t_copula_pipeline(df_high, cols=("孕周","孕妇BMI","Y染色体浓度"), do_plot=True, sim_n=100)
fitres = res['fit']        # 拟合结果
data_array = res['data_array']  # 原始数据数组
arr_sorted_list = [np.sort(data_array[:, j]) for j in range(data_array.shape[1])]  # 排序列表


y_threshold = 0.04         # 你要判定的 Y 浓度阈（根据你数据的量纲设定）
lam = 0.5                  # 惩罚系数（你可以调）
bmi_fixed = 37.5           # 在 BMI=40 的孕妇上优化时点
res = find_optimal_t(fitres, data_array, arr_sorted_list, y_threshold, lam=lam, bmi_fixed=bmi_fixed,
                     t_grid=np.linspace(10,27,69), n_cond_draw=3000)
print("Best t:", res['best'])
#
# # 可视化 U(t) 与 prob(t)
xs = np.array([r['t'] for r in res['grid_results']])
us = np.array([r['mean_U'] for r in res['grid_results']])
ps = np.array([r['mean_prob'] for r in res['grid_results']])

# 生成更密的 x 以便平滑
xs_smooth = np.linspace(xs.min(), xs.max(), 300)

# 三次样条拟合
spl_prob = make_interp_spline(xs, ps, k=3)
ps_smooth = spl_prob(xs_smooth)

spl_U = make_interp_spline(xs, us, k=3)
us_smooth = spl_U(xs_smooth)

# 绘图
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.plot(xs_smooth, ps_smooth, color='blue')
#plt.scatter(xs, ps, color='red', label='original points')
plt.xlabel('gestational weeks')
plt.ylabel('Probability')
plt.title('P(Y >= 0.04 | t,BMI)')
plt.legend()

plt.subplot(1,2,2)
plt.plot(xs_smooth, us_smooth, color='green')
#plt.scatter(xs, us, color='red', label='original points')
plt.xlabel('gestational weeks')
plt.ylabel('U(t)')
plt.title('U(t)=prob - lam * C(t)')
plt.legend()

plt.tight_layout()
plt.show()
```
