```python
import pandas as pd
df = pd.read_excel('ref.xlsx')
print(df.head())
```

```python
age = df['年龄']
print (len(df))
```

```python
df = df[df['原始读段数'] >= 3000000]
df = df[df['在参考基因组上比对的比例'] <= 0.9]
df = df[df['在参考基因组上比对的比例'] >= 0.7]
df = df[df['重复读段的比例'] <=0.2]
df = df[df['唯一比对的读段数  '] >=2900000]
df = df[df['GC含量'] >=0.38]
df = df[df['GC含量'] <=0.46]
df = df[df['被过滤掉读段数的比例'] <=0.25]
print (len(df))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

from pylab import mpl
# 设置中文显示字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
mpl.rcParams["axes.unicode_minus"] = False

# 设置风格
sns.set_style("whitegrid")
plt.figure(figsize=(15, 10))

# 选择您过滤的指标
qc_metrics = ['原始读段数', '在参考基因组上比对的比例', '重复读段的比例', '唯一比对的读段数  ', 'GC含量', '被过滤掉读段数的比例']
titles = ['Raw Reads', 'Mapping Rate', 'Duplication Rate', 'Uniquely Mapped Reads', 'GC Content', 'Filtered Reads Ratio']

# 创建子图
for i, (col, title) in enumerate(zip(qc_metrics, titles), 1):
    plt.subplot(2, 3, i)
    
    # 绘制小提琴图展示分布
    sns.violinplot(y=df[col], color="skyblue", inner=None)
    sns.swarmplot(y=df[col], color="black", size=3, alpha=0.5) # 显示每个样本点
    
    # 根据不同的指标添加阈值线
    if col == '在参考基因组上比对的比例':
        plt.axhline(y=0.7, color='red', linestyle='--', label='Min Threshold (0.7)')
        plt.axhline(y=0.9, color='red', linestyle='--', label='Max Threshold (0.9)')
    #elif col == '重复读段的比例':
    #    plt.axhline(y=0.05, color='red', linestyle='--', label='Threshold (0.2)')
    # ... 为您其他的阈值添加参考线
    
    plt.title(f'Distribution of {title}')
    if i == 1: # 只在第一个图显示图例
        plt.legend()

plt.tight_layout()
plt.show()
```
